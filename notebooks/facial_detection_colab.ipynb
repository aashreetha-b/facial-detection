```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt
import numpy as np

# Load LFW dataset with min_faces_per_person=50 for better quality samples
lfw_people = fetch_lfw_people(min_faces_per_person=50, resize=0.4)

print(f"Total images: {lfw_people.images.shape[0]}")
print(f"Image shape: {lfw_people.images[0].shape}")
print(f"Number of classes: {len(lfw_people.target_names)}")
```

```python
fig, axes = plt.subplots(3, 5, figsize=(15, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(lfw_people.images[i], cmap='gray')
    ax.set_title(lfw_people.target_names[lfw_people.target[i]])
    ax.axis('off')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Flatten images (each image to 1D array)
X = lfw_people.images.reshape(len(lfw_people.images), -1)
y = lfw_people.target

# Split 75% train, 25% test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

print(f"Training samples: {X_train.shape[0]}")
print(f"Test samples: {X_test.shape[0]}")
```
